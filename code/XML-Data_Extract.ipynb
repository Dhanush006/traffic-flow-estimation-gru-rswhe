```python
dir_name =  os.path.abspath('/content/gdrive/MyDrive/IT-CV/Datasets/UA-DETRAC/Train/ANNOTS/MVI_20034/')
list_of_files = sorted( filter( lambda x: os.path.isfile(os.path.join(dir_name, x)),
                        os.listdir(dir_name) ) )
list_of_files = natsort.natsorted(list_of_files)
# list_of_files = list_of_files[0:400]
l1 = []
for i in list_of_files:
  with open('/content/gdrive/MyDrive/IT-CV/Datasets/UA-DETRAC/Train/ANNOTS/MVI_20034/' + i, 'r') as f:
    data = f.read()    
  Bs_data = BeautifulSoup(data, "xml")
  b_unique = Bs_data.find_all('box')
  l2 = []
  for i in b_unique:
    top = float(i.get('top'))
    height = float(i.get('height'))
    # dist = float("{:.0f}".format(250 - ((top) + height/2)))
    dist = float("{:.2f}".format(((top + height/2) - 250)))
    # ymin = int(i.find('ymin').contents[0])
    # ymax = int(i.find('ymax').contents[0])
    # dist = float(100 - (ymin + ymax)/2)
    l2.append(dist)
  l2 = sorted(l2, key=abs)
  l2 = l2[0:3]
  # l2 = sorted(l2)
  l1.append(l2)
```

```python
def sgn(val):
  if(val>0):
    return 1
  elif(val<0):
    return -1
  else:
    return 0
def categ(cl):
  if(cl>0):
    return 2
  elif(cl<0):
    return 3
  else:
    return 1
```

```python
t = []
for i in range(1,len(l1)):
  sum = 0
  for j in range(0,3):
    if(abs(l1[i][j]-l1[i-1][j])<5):
      sum +=  sgn(l1[i][j]) - sgn(l1[i-1][j])
    
  t.append(categ(sum))
```
